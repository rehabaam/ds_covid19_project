```python
from src.preprocessing.image_preprocessor import get_images_statistics_by_scales,plot_images_statistics
import pandas as pd

# Available categories in the dataset
dataset = ['COVID','NORMAL','Viral Pneumonia','Lung_Opacity']

# Path to the raw data and preprocessed data
raw_data_dir = '../data/raw/COVID-19_Radiography_Dataset/'
preprocessed_data_dir = '../data/preprocessed/'
```

## Plotting image statistics by scales (Normalization vs. standardization)

```python
for i in dataset:
    normalized_images, standardized_images = get_images_statistics_by_scales(raw_data_dir+i+'/images')
    for col in ['mean','std','min','max']:
        data = pd.DataFrame().assign(normalized=normalized_images[col],standardized=standardized_images[col])
        plot_images_statistics(f'{i} {col}',data)
    del normalized_images, standardized_images
```
